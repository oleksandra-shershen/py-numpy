### Task 7.2

````python
# Python code
def unknown_signature(lst1, lst2):
    bitwise_complement = [~x for x in lst1]
    bitwise_and = [x & y for x, y in zip(lst1, lst2)]
    bitwise_or = [x | y for x, y in zip(lst1, lst2)]
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or
````

````python
# write your numpy code here
def unknown_signature_np(lst1, lst2):
    arr1 = np.array(lst1)
    arr2 = np.array(lst2)
    
    bitwise_complement = np.bitwise_not(arr1)
    bitwise_and = np.bitwise_and(arr1, arr2)
    bitwise_or = np.bitwise_or(arr1, arr2)
    
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
result_bitwise_complement_np, result_bitwise_and_np, result_bitwise_or_np = unknown_signature_np(list1, list2)

result_bitwise_complement_np, result_bitwise_and_np, result_bitwise_or_np
````

### Task 7.3 Optional

````python
# Python code
def unknown_signature(lst):
    result = []
    for num in lst:
        if num % 2 == 0 and (num % 3 == 0 or num > 10):
            modified_num = (num ^ 7) + 2
            result.append(modified_num)
    return result

my_list = [6, 9, 12, 14, 17, 20]
result_python_style = unknown_signature(my_list)
result_python_style
````

````python
# write your numpy code here
def unknown_signature(arr):
    mask = (arr % 2 == 0) & ((arr % 3 == 0) | (arr > 10))
    modified_arr = ((arr ^ 7) + 2)[mask]
    return modified_arr

my_array = np.array([6, 9, 12, 14, 17, 20])
unknown_signature(my_array)
````
